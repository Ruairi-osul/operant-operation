```python
from op_op.load import load_motion
import os
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_DIR = Path(os.environ.get("DATA_DIR")) / "raw"

sns.set_context("notebook", font_scale=1.5)
```

```python
MOUSE = "example-mouse"
SESSIONS = (
    "rm-early",
    "rm-late",
    "rdt-d1",
    "rdt-d2",
    "rdt-d3",
)
SESSION = "rdt-d1"
```

```python
df = load_motion(DATA_DIR, MOUSE, SESSION)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 3), sharex=True)

ax1.plot(df["time"], df["x"], label="x", color="purple", alpha=0.5)
ax1.plot(df["time"], df["y"], label="x", color="orange", alpha=0.5)

ax1.set_ylabel("Mouse Position (px)")

ax2.plot(df["time"], df["velocity"], color="black")
ax2.set_ylabel("Mouse Velocity (px/s)")


ax2.set_xticklabels([int(x / 60) for x in ax2.get_xticks().tolist()], fontsize=12)

ax2.set_xlabel("Time (min)")

ax1.yaxis.set_visible(False)
ax2.yaxis.set_visible(False)

sns.despine(fig=fig, left=True, bottom=True)
```
